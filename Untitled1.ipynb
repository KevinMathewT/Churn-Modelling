```python
def nonlin(x, derivative = False):
    if(derivative == True):
        return (x * (1 - x))
    else:
        return (1 / (1 + np.exp(-x)))
```

```python
learning_rate = 0.01
input_size = X.shape[1]
hidden_size = 16
output_size = 1
```

```python
Wxh = np.random.randn(hidden_size, input_size).astype(float) * (0.001)
Who = np.random.randn(output_size, hidden_size).astype(float) * (0.001)
bh = np.zeros((hidden_size, 1)).astype(float)
bo = np.zeros((output_size, 1)).astype(float)
nonlin(bh+2)
```

```python
def forward_propagation(inputs):
    hidden = nonlin(Wxh.dot(inputs), False).reshape(-1, 1)
    outputs = nonlin(Who.dot(hidden), False).reshape(-1, 1)
#     print(inputs, '-------------')
#     print(Wxh.dot(inputs), '---')
#     print('hidden : \n------\n')
#     print(hidden)
    outputs[0, 0] = min(max(outputs[0, 0], 1 - 1e-10), 1e-10)
#     print(outputs)
#     print(hidden.shape)
#     print(outputs.shape)
    return hidden, outputs
```

```python
def backpropagation(inputs, hidden, outputs, y):
    loss = (y * np.log(outputs)) + ((1 - y) * np.log(1 - outputs))
    print(loss)
    del_output = (((y / outputs) + ((y - 1) / (1 - outputs))) * (nonlin(outputs, True)))
    del_hidden = (Who.T.dot(del_output)) * (nonlin(hidden, True))
    
    return del_output, del_hidden
```

```python
for i in range(X_train.shape[0]):
    inputs = X_train[i].reshape(-1, 1)
    y = y_train[i]
    hidden, outputs = forward_propagation(X_train[i])
    del_output, del_hidden = backpropagation(inputs, hidden, outputs, y)
    inputs.shape
    Wxh = Wxh - learning_rate * (del_hidden.dot(inputs.T))
    Who = Who - learning_rate * (del_output.dot(hidden.T))
```
