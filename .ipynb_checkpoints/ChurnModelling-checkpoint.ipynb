```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = np.array(pd.read_csv('Churn_Modelling.csv'))
data = data
data
```

```python
X = data[:, 3:-1]
y = data[:, -1]
print(X)
print(y)
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, StandardScaler
encoder_1 = LabelEncoder()
X[:, 1] = encoder_1.fit_transform(X[:, 1])
encoder_2 = LabelEncoder()
X[:, 2] = encoder_2.fit_transform(X[:, 2])
onehotencoder = OneHotEncoder(categorical_features = [1])
X = onehotencoder.fit_transform(X).toarray()
X = X[:, 1:]
scaler = StandardScaler()
X = scaler.fit_transform(X)
X = X.astype(float)
y = y.astype(float)
print(X)
y = y.reshape(-1, 1)
X_test = X[int(X.shape[0]*0.8):, :]
X = X[:int(X.shape[0]*0.8), :]
y_test = y[int(y.shape[0]*0.8):, :]
y = y[:int(y.shape[0]*0.8), :]
print(str(X.shape) + " " + str(y.shape))
```

```python
learning_rate = 0
number_of_inputs = X.shape[0]
number_of_training_examples = int(X.shape[0]*0.8)
input_size = X.shape[1]
hidden_size = 5
output_size = 1
epochs = 1
```

```python
Wxh = np.random.randn(hidden_size, input_size).astype(float) * 0.01
Who = np.random.randn(output_size, hidden_size).astype(float) * 0.01
bh = np.zeros((hidden_size, 1)).astype(float)
bo = 0.0
print(Wxh)
```

```python
def sigmoid(x, derivative = False):
    if(derivative):
        return (x*(1-x))
    return 1 / (1 + np.exp(-x))
```

```python
def forward_pass(inputs):
    hidden = sigmoid(Wxh.dot(inputs) + bh)
    output = sigmoid(Who.dot(hidden) + bo)
    output = min(max(output, 1e-15), 1-1e-15)
    return hidden, output
```

```python
def backpropagation(inputs, hidden, output, y):
    if y[0] == 0:
        loss = -np.log(1 - output)
    else:
        loss = -np.log(output)
    print(loss.shape)
    global Wxh, Who, bh, bo
    
    if(y[0] == 0):
        del_output = (((1 - output))) * sigmoid(Who.dot(hidden) + bo, True)
    else:
        del_output = (((output))) * sigmoid(Who.dot(hidden) + bo, True)
    del_hidden = (Who.T.dot(del_output) * sigmoid(Wxh.dot(inputs) + bh, True))
    
    return del_output, del_hidden
```

```python
mxh = np.zeros_like(Wxh)
mho = np.zeros_like(Who)
mbo = np.zeros_like(bo)
mbh = np.zeros_like(bh)
print(mbo.shape)
for qwerty in range(epochs):
    for i in range(number_of_inputs):
        global Wxh, Who, bh, bo, mxh, mho, mbo, mbh
        hidden, output = forward_pass(X[i].reshape(-1, 1))
        del_output, del_hidden = backpropagation(X[i].reshape(-1, 1), hidden, output, y[i])
        Who_derivative = del_output.dot(hidden.T)
        Wxh_derivative = del_hidden.dot(X[i].reshape(-1, 1).T)
#         print(del_output)
#         print(del_hidden)
        
        mxh += Wxh_derivative * Wxh_derivative
        mbh += del_hidden * del_hidden
        mho += Who_derivative * Who_derivative
        mbo += del_output[0][0] * del_output[0][0]
#         print(mbo)
        Wxh = Wxh - (learning_rate * Wxh_derivative) / np.sqrt(mxh + 1e-8)
        bo = bo - learning_rate * del_output / np.sqrt(mbo + 1e-8)
        Who = Who - learning_rate * Who_derivative / np.sqrt(mho + 1e-8)
        bh = bh - learning_rate * del_hidden / np.sqrt(mbh + 1e-8)
```

```python
for i in range(number_of_training_examples):
    hidden, output = forward_pass(X_test[i].reshape(-1, 1))
    print(str(output) + " - " + str(y_test[i]))
```
